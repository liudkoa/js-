### 开始js解密

#### 从主页拿到__signature参数，这是直接从后端传过派来的

```python
url='http://zwfw.san-he.gov.cn/icity/icity/guestbook/interact#'
response=requests.get(url=url)
print(response.cookies)
print(response.status_code)
data=response.text
```

```python
import re
__signature = re.search(r'var __signature = "(.*?)";',data,flags=re.S).group(1)
chars = "0123456789abcdef"
```

#### 解决s

```python
from time import time
import math
import numpy as np
a=int(np.random.random()*(9999 - 1000 + 1) + 1000)
print(a)
b=math.floor(time())*1000
print(b)
curTime=str(a)+str(b)
print(curTime)

d=int(np.random.random()* (15 - 15 + 1) + 10)
print(d)
e=chars[d]
print(e)
s= e+curTime
print(s)
#前面一大堆都是分析第一个s的过程,现在不用了
s=__signature 
print(s)
```
